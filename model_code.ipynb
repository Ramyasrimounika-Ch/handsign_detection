```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("merged.csv")
```

```python
X = df.drop('label', axis=1).values  # shape: (num_samples, 42)
y = df['label'].values
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(42,)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(4, activation='softmax')  # 5 classes: A, B, C, R, 4
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test), batch_size=32)
```

```python
prediction_lastlayer=model.predict(X_test)
```
